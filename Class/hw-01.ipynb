###  6(50). 상기 "alba.csv"  를 가공하여, 다음  표와 같은 Dataframe 를 만들어, 출력하고, 또한 이를 'alba.html' 파일로 저장해보시오.

![Alt text](알바.png)

```python
df = pd.read_csv('alba.csv',encoding="euc-kr")
type(df)
raw_date = {'구':['강남구', '강동구', '강북구', '강서구', '관악구', ' 광진구', ' 구로구',' 금천구', ' 노원구', ' 도봉구', '동대문구', '동작구', '마포구', '서대문구','서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구',' 은평구', '종로구', '중구', '중랑구'],
            '알바수':df.groupby('area1')['hourly_wage'].count(),
            '시급평균':df.groupby('area1')['hourly_wage'].mean(),
            '시급최고':df.groupby('area1')['hourly_wage'].max(),
            '시급최저':df.groupby('area1')['hourly_wage'].min()
           }
           
df = pd.DataFrame(raw_date, columns = ['구', '알바수', '시급평균', '시급최고', '시급최저'])
df.reset_index(inplace=True)
df.drop(columns = ['area1'],inplace=True)
df
```

```python
df.to_html('alba.html')
```
