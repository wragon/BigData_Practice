```python
import seaborn as sns
import pandas as pd

iris= sns.load_dataset("iris")  #sample dataframe 불러오기
iris.to_csv("iris.csv", index=False)  #save
```

```python
import matplotlib.pyplot as plt

iris.species.value_counts().plot(kind='pie')
```

```python
sns.countplot(x='sepal_length', data=iris)
```

```python
sns.histplot(x='sepal_length', data=iris)  #bin 자동설정
```

```python
sns.histplot(x='sepal_length', data=iris, bins=[3.5,4.5,5.5,6.5,7.5,8.5]) # 또는 bins= 갯수
```

```python
sns.histplot(x='sepal_length', data=iris, bins=[4, 5, 6, 7, 8]) # 또는 bins= 갯수
```

```python
# Iris Pair Plot, Hue로 꽃의 종을 시각화
sns.pairplot(iris, hue="species", markers=["o", "s", "D"],
             palette="husl")
plt.show()
```

```python
ax = plt.subplots()
ax = sns.lineplot(x='species', y='sepal_width', data=iris)
ax.set_title('iris')
ax.set_xlabel('species')
ax.set_ylabel('sepal_width')
```
