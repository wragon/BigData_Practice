```python
from konlpy.tag import Okt
from konlpy.tag import Kkma
from collections import Counter
```

```python
#-*- coding: utf-8 -*-
news = ''
news_list = []
for i in range(10):
    fd = open(str(i) + ".txt", "rt", encoding='UTF8')
    lines= fd.read()
    fd.close()
    news = news + lines
    news_list.append(lines)

news_list
```

```python
nlp = Okt()
message_N = nlp.nouns(news)
message_N  
```

```python
count = Counter(message_N)
count 
```

```python
word_count = dict()

for tag, counts in count.most_common(60):
    if(len(str(tag))>1):
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
```

```python
len(word_count)
```
